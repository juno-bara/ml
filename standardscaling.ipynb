```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

iris = load_iris()
data = iris.data
iris_df = pd.DataFrame(data, columns=iris.feature_names)

iris_df.describe()
```

```python
iris_df
```

```python
plt.subplot(2,2,1).set_title("sepal length (cm)")
plt.scatter(iris_df.index,iris_df["sepal length (cm)"])

plt.subplot(2,2,2).set_title("sepal width (cm)")
plt.scatter(iris_df.index,iris_df["sepal width (cm)"])

plt.subplot(2,2,3).set_title("petal length (cm)")
plt.scatter(iris_df.index,iris_df["petal length (cm)"])

plt.subplot(2,2,4).set_title("petal width (cm)")
plt.scatter(iris_df.index,iris_df["petal width (cm)"])

plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['target'])
```

```python
plt.scatter(df['sepal length (cm)'],df['sepal width (cm)'],alpha=0.4,s=100*df['petal width (cm)'],c=df['target'],cmap=plt.cm.get_cmap('prism', df['target'].nunique()),edgecolors='black')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
cb=plt.colorbar(label='species')
cb.set_ticks([0,1,2])
cb.set_ticklabels(["setosa","versicolor","virginica"])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# StandardScaler객체 생성
scaler = StandardScaler()
# StandardScaler로 데이터 셋 변환 , fit()과 transform()호출
iris_scaled = scaler.fit_transform(iris_df)

# transform()시 scale 변환된 dataset이 numpy ndarry로 반환되어 이를 DataFrame으로 변환
iris_df_scaled = pd.DataFrame(data=iris_scaled,columns=iris.feature_names)
print('feature들의 평균 값')
print(iris_df_scaled.mean(),'\n')

print('feature들의 분산 값')
print(iris_df_scaled.var())

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title("sepal length (cm)")
plt.scatter(iris_df_scaled.index,iris_df_scaled["sepal length (cm)"])
plt.subplot(2,2,2)
plt.title("sepal width (cm)")
plt.scatter(iris_df_scaled.index,iris_df_scaled["sepal width (cm)"])
plt.subplot(2,2,3)
plt.title("petal length (cm)")
plt.scatter(iris_df_scaled.index,iris_df_scaled["petal length (cm)"])
plt.subplot(2,2,4)
plt.title("sepal width (cm)")
plt.scatter(iris_df_scaled.index,iris_df_scaled["petal width (cm)"])
plt.show()
```
